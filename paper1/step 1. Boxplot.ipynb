```python
#run here for boxplot.
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
data = pd.read_excel('\\Data\\Data_belisario2.xlsx', 'Sheet1')
data= data[data['FECHA_UNIDAD']>='2016-01-01 00:00:00']
data= data[data['FECHA_UNIDAD']<'2020-03-16 00:00:00']

#data
data_train = data[data['FECHA_UNIDAD']<'2020-01-16 00:00:00']
data_test = data[data['FECHA_UNIDAD']>= '2020-01-16 00:00:00']

data_train["Dataset"]="Train"
data_test["Dataset"]="Test"

data1 = pd.concat([data_train,data_test])
data1.columns = ["Date","CO","NO2","SO2","PM2.5","O3","Temperature","SR","p","Prec","RH","WD","WS","Dataset"]
#FECHA_UNIDAD 	CO_mg/m3 	NO2_ug/m3 	SO2_ug/m3 	PM2.5_ug/m3 	O3_ug/m3 	Values 	T_°C 	SR_W/m2 	p_mb 	Prec_mm 	RH_% 	WD_° 	WS_m/s
                 
#Inputs = data1.iloc[:,6:]
#Inputs.columns = ["Values","SR","p", Prec_mm 	RH_% 	WD_° 	WS_m/s]
#Outputs = data1.iloc[:,1:7]
#Outputs.columns = ["CO","NO2","SO2","PM2.5","O3","Values"]
data1
```

```python
import matplotlib.pyplot as plt
data1['CO'] = pd.to_numeric(data1['CO'], errors = 'coerce')
data1['CO']= data1['CO'].abs()
data1['NO2'] = pd.to_numeric(data1['NO2'], errors = 'coerce')
data1['NO2']= data1['NO2'].abs()
data1['SO2'] = pd.to_numeric(data1['SO2'], errors = 'coerce')
data1['SO2']= data1['SO2'].abs()
data1['PM2.5'] = pd.to_numeric(data1['PM2.5'], errors = 'coerce')
data1['PM2.5']= data1['PM2.5'].abs()
data1['O3'] = pd.to_numeric(data1['O3'], errors = 'coerce')
data1['O3']= data1['O3'].abs()

columns = ["Dataset","CO","NO2","SO2","PM2.5","O3","Temperature","SR","p","Prec","RH","WD","WS"]
columns2 = ["CO","NO2","SO2","PM2.5","O3","Temperature","SR","p","Prec","RH","WD","WS"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
myFontSize= 18
plt.figure()
fig, axs = plt.subplots(3,3,figsize=(14,17))
#columns2 = ["CO","NO2","SO2","PM2.5","O3","Temperature","SR","p","Prec","RH","WD","WS"]
#CO_mg/m3 	NO2_ug/m3 	SO2_ug/m3 	PM2.5_ug/m3 	
#O3_ug/m3 	Values 	T_°C 	SR_W/m2 	p_mb 	Prec_mm 	RH_% 	WD_° 	WS_m/s

b= sns.boxplot(x='Dataset', y='SR', data=data1, ax=axs[0,0],width=0.3)
b.set_xlabel("a) SR (W/m2)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)

b = sns.boxplot(x='Dataset', y='WD', data=data1, ax=axs[0,1],width=0.3)
b.set_xlabel("b) WD (°)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)

b=sns.boxplot(x='Dataset',y='WS', data=data1, ax=axs[0,2],width=0.3)
b.set_xlabel("c) WS (m/s)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)


b=sns.boxplot(x='Dataset', y='Prec', data=data1, ax=axs[1,0],width=0.3)
b.set_xlabel("d) Prec (mm)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)

b=sns.boxplot(x='Dataset',y='p', data=data1, ax=axs[1,1],width=0.3)
b.set_xlabel("e) p (mb)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)

b=sns.boxplot(x='Dataset',y='Temperature', data=data1, ax=axs[1,2],width=0.3)
b.set_xlabel("f) Temperature (°C)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)


b=sns.boxplot(x='Dataset',y='RH', data=data1, ax=axs[2,1],width=0.3)
b.set_xlabel("g) RH (%)",fontsize=myFontSize)
b.set_ylabel("",fontsize=myFontSize)
b.tick_params(labelsize=myFontSize)

axs[2, 0].remove()
axs[2, 2].remove()
#plt.savefig('Meteorological features.svg', format='svg')
```
