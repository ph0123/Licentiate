```python


import pandas as pd
import plotly
import psutil
import requests
import orca

def format_time(PreLD, column_name):
    PreLD[column_name] = pd.to_datetime(PreLD[column_name], format='%Y-%m-%d %H:%M:%S')
    PreLD['Date'] = PreLD[column_name].dt.strftime('%Y-%m-%d')
    first_col = PreLD.pop('Date')
    PreLD.insert(0, 'Date', first_col)
    return PreLD.groupby(['Date']).mean()


def Quatifying(target, inputfile):
    print(inputfile, '...')
    data =  pd.read_csv(inputfile)
    
    LD= data[data['DateTime']<= '2020-06-01 23:00:00']
    LD= LD[LD['DateTime']> '2020-03-15 23:00:00']
    PL= data[data['DateTime']> '2020-06-01 23:00:00']

    Sum_LD_real = LD[target].sum()
    Sum_LD_pred = LD['Predicted'].sum()
    print('percentages of LD: ',(Sum_LD_real-Sum_LD_pred)*100/(Sum_LD_pred),'%')

    Sum_PL_real = PL[target].sum()
    Sum_PL_pred = PL['Predicted'].sum()
    print('percentages of PL: ',(Sum_PL_real-Sum_PL_pred)*100/(Sum_PL_pred),'%')
    
    

Quatifying('PM2.5_ug/m3',"PM2_5_Results_Quatifying.csv" )
Quatifying('SO2_ug/m3',"SO2_Results_Quatifying.csv" )
Quatifying('O3_ug/m3',"O3_Results_Quatifying.csv" )
Quatifying('NO2_ug/m3',"NO2_Results_Quatifying.csv" )
Quatifying('CO_mg/m3',"CO_Results_Quatifying.csv" )
```
