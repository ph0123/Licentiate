```python
import pandas as pd
import numpy as np


def get_min_max_avg(Train_Path_List,Test_Path_List): #train and test
    Outputs_Train = pd.DataFrame()
    Outputs_Test = pd.DataFrame()
    for Train_Path,Test_Path in zip(Train_Path_List, Test_Path_List):
        Train = pd.read_csv(Train_Path)
        Test  = pd.read_csv(Test_Path)
        Train = Train.iloc[:,:-1]
        Test = Test.iloc[:,:-1]
        Outputs_Train = pd.concat([Outputs_Train,Train])
        Outputs_Test = pd.concat([Outputs_Test,Test])
        break
    return [Outputs_Train.min(),Outputs_Train.max(),Outputs_Train.sum()/(Outputs_Train.shape[0]),Outputs_Train.std()], [Outputs_Test.min(),Outputs_Test.max(),Outputs_Test.sum()/(Outputs_Test.shape[0]), Outputs_Test.std()]
#return_list = [] // min, max, average
```

```python



#name= all_data.col
#Name= ['Index','Hours','Julian Day','Week Day','Temperature','SR','p','Prec','RH','WD','WS']
Names = ['min', 'max', 'avg', 'std']
data_path= 'C:\\data\\PeakConcentration-Idea2-2-Clasification\\FullDay_Peak_Values\\'
Train_Path_List = [data_path+'NO2_Train.csv',data_path+'CO_Train.csv',data_path+'O3_Train.csv',data_path+'PM2_5_Train.csv', data_path+'SO2_Train.csv']
Test_Path_List = [data_path+'NO2_Test.csv', data_path+'CO_Test.csv', data_path+'O3_Test.csv', data_path+'PM2_5_Test.csv', data_path+'SO2_Test.csv']
Train_, Test_ = get_min_max_avg(Train_Path_List, Test_Path_List)

df1 = pd.DataFrame(Train_)
df1.insert(0, 'measurements_train', Names)
df2 = pd.DataFrame(Test_)
df2.insert(0, 'measurements_test', Names)

with open('statistic.csv','a') as f:
    f.write("Training Data")
    df1.to_csv(f, index=False)
    #f.write("\n")
    f.write("Testing Data")
    df2.to_csv(f, index=False)

print("Done to file statistic.csv")
```
